# Data preporation

In this notebook I make transformations in data
1. Extract a domain from email
2. Enrich dataset with data about countries


```python
import numpy as np
import pandas as pd
from IPython.display import HTML
```

```python
df = pd.read_csv("../data/car_purchasing.csv",encoding='ISO-8859-1')
df.columns = df.columns.str.replace(" ", "_").str.replace("-","").str.lower()
```

```python
df.head()
```

# Extract domain from email

```python
x = "dsfsdf.sdfsdf.sdfsd.f"
x[x.find("."):]
```

```python
df.customer_email.str.split('@').apply(lambda x: x[1][x[1].find('.'):])
```

```python
df["customer_email_suffix"] = df.customer_email.str.split('@').apply(lambda x: x[1][x[1].find('.'):])
```

# Enrich with countries data

I've got data about countries from here: https://www.kaggle.com/datasets/fernandol/countries-of-the-world

```python
countries_df = pd.read_csv("../data/countries_of_the_world.csv")
```

```python
countries_df.head()
```

```python
countries_df.columns = countries_df.columns\
    .str.replace(" ", "_")\
    .str.replace("[-|(|)|\.|//|\$|\%]","", regex=True)\
    .str.lower()
```

```python
countries_df.country = countries_df.country.str.strip()
```

```python
df.country = df.country\
    .str.replace("Bahamas", "Bahamas, The")\
    .str.replace("Central African Republic", "Central African Rep.")\
    .str.replace("Falkland Islands", "Argentina")\
    .str.replace("Macao", "Macau")\
    .str.replace("Micronesia", "Micronesia, Fed. St.")\
    .str.replace("Montenegro", "Serbia")\
    .str.replace("Trinidad and Tobago", "Trinidad & Tobago")\
    .str.replace("United Kingdom (Great Britain)", "United Kingdom", regex=False)\
    .str.replace("United States Minor Outlying Islands", "United States")\
    .str.replace("Viet Nam", "Vietnam")\
    .str.replace("Virgin Islands, British", "United Kingdom")\
    .str.replace("Virgin Islands, United States", "United States")\
.str.replace("Bosnia and Herzegovina", "Bosnia & Herzegovina")\


    
```

```python
all_country = set(countries_df.country)
all_country
```

# Remove observation from unknown countries

```python
not_found_countries = set(df[~ df.country.isin(all_country)].country)
```

```python
not_found_countries
```

```python
df = df[df.country.isin(all_country)]
```

```python
df.shape
```

```python
df_enriched = df.merge(countries_df, how='left', on='country')
```

```python
df_enriched.head()
```

```python
df_enriched.info()
```

# Let's fill NA

```python
na_values = df_enriched.isna().sum()
```

```python
na_values
```

Let's look to these obs.

```python
def show_na_obs(column):
    display(HTML("<h3>" + column + "</h3>"))
    display(df_enriched[column].describe())
    display(df_enriched[df_enriched[column].isna()])
```

```python
for column in na_values[na_values > 0].index:
    show_na_obs(column)
```

```python
df_enriched.fillna(0, inplace=True)
df_enriched.info()
```

```python
df_enriched.to_csv("../data/enriched_data.csv")
```
